```python
Flowers = Path("flowers")
```

```python
transformaciones = fv.get_transforms(flip_vert=True, max_rotate=360,
                         max_zoom=1.25, max_lighting=0.3)
```

```python
data = (fv.ImageDataBunch
        .from_folder(Flowers, ds_tfms=transformaciones, size=224, bs=64) #bs = batch_size
        .normalize(fv.imagenet_stats))
```
